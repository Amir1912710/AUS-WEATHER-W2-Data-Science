```python
import requests
import json

# API URL
url = "https://www.banquemisr.com/api/sitecore/LocationsApi/GetPlaces?GovernateId=85616bbc-0678-4f4a-9947-63f367caff47&AreaId=00000000-0000-0000-0000-000000000000&IsBranch=false&IsAtm=true&lang=en"

# Send an HTTP GET request to the API
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    data = json.loads(response.text)
    
    # Extract and print the branch location details
    for location in data:
        location_name = location.get("name")
        location_address_lat = location.get("lat")
        location_address_lng = location.get("lng")
        print("Location Name:", location_name)
        print("Location Lat:", location_address_lat)
        print("Location Lat:", location_address_lng)
        print("=" * 30)

else:
    print("Failed to retrieve data from the API. Status code:", response.status_code)
```
