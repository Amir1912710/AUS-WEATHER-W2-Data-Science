```python
data=data.drop(['Evaporation','Sunshine', 'Cloud9am', 'Cloud3pm'],axis=1)
data.head()
```

```python
data=data.dropna(axis=0)
data.shape
```

```python
data.isnull().sum()
```

```python
catcols=data.select_dtypes(include='object').columns
for i in catcols:
    print(i)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in catcols:
    data[i]=le.fit_transform(data[i])
```

```python
data.dtypes
```

```python
X=data.drop(columns=["RainTomorrow"],axis=1)
y=data["RainTomorrow"]
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='MaxTemp',y='MinTemp',hue='RainTomorrow',palette='inferno',data=data);
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),annot=True);
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='Humidity9am',y='Temp9am',hue='RainTomorrow',palette='inferno',data=data);
```

```python
#
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix ,accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
score=rf.score(X_test,y_test)
print(f'Model accuracy on test ={score*100:.2f}%')
```

```python
y_pred=rf.predict(X_test)
print(f"Classification report:\n\n {classification_report(y_test,y_pred)}")
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_train,y_train)
print('before resampling')
print(y_train.value_counts())
print('\n')
print('after using smote')
print(pd.Series(y_resampled).value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf=RandomForestClassifier()
rf.fit(X_resampled,y_resampled)
```

```python
score=rf.score(X_test,y_test)
print(f'Model accuracy on test ={score*100:.2f}%')
```
